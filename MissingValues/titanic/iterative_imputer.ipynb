### Filing missing values with IterativeImputer  
#### (It applies only on Numeric Data)  
It iterate over complete dataset for the repititon we mention and than calculate mean / median of every iteration and gets the perfect value to impute  
To use this first enable multiple iterative  `from sklearn.experimental import enable_iterative_imputer` and than import the specfic imputer `from sklearn.impute import IterativeImputer`
Assign number of iterations on initialization of imputer  
use fit_transform to impute the null values  
`YEH COMPLETE ROW P ML KA ALGO LAGA K MISSING VALUE KO  PREDICT KRTA HA`


```python
# importing libraries

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

df = sns.load_dataset('titanic')
imp = IterativeImputer(max_iter=10)
df['age'] = imp.fit_transform(df[['age']])
df['age'].isnull().sum()
df['age'].head()

```

### Filling missing values in Multiple features

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder

dcat = sns.load_dataset('titanic')
le = LabelEncoder()
cols = {"embarked" , "class" , "who" , "embark_town" , "alive" , "deck" , "sex"}
for c in cols :
    dcat[c] = le.fit_transform(df[c])


ip = IterativeImputer(max_iter=12)

for c in cols:
    dcat[c] = ip.fit_transform(dcat[[c]])

dcat.head()

sns.histplot(data=dcat , x="age" , hue="who" , kde="true")
```

### Label Encoding (Single Feature)

### Multiple Features Encoding

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

cols = {"embarked" , "class" , "who" , "embark_town" , "alive" , "deck"}
le = LabelEncoder()

for c in cols:
    df[c] = le.fit_transform(df[c])
    
impu = IterativeImputer()
for col in cols:
    df[col] = impu.fit_transform(df[[col]])
    
df['deck'].head()
```
