# Feature Transformation
Feature transformation is a crucial step in the data preprocessing pipeline that involves converting raw data into a format that is more suitable for analysis and modeling. This process can enhance the performance of machine learning algorithms by improving the quality of the input features. It can also change the distribution of the data, making it easier to identify patterns and relationships.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
df = sns.load_dataset('titanic')
impt = KNNImputer(n_neighbors=5)
df['age'] = impt.fit_transform(df[['age']])
sns.histplot(df['age'] , kde=True)
```

```python
from sklearn.preprocessing import PowerTransformer

new_df = sns.load_dataset('titanic')
trans = PowerTransformer(method="yeo-johnson")
plt.title("Fare Feature Befor Transformation")
sns.histplot(new_df['fare'] , kde=True)
plt.show()
```

```python
new_df['fare'] = trans.fit_transform(new_df[['fare']])
plt.title("Fare Feature after Yeo Johonson Transformation")
sns.histplot(new_df['fare'] , kde=True)
plt.show()
```

# BOX COX
This method is used only for positive data. where values are greater than 0. If we have negative values or zero values then we can not use this method. Instead we can use Yeo Johnson method / Quantile.

```python
new_df = sns.load_dataset('titanic')
trans = PowerTransformer(method="box-cox")

# new_df['fare'] = trans.fit_transform(new_df[['fare']])
plt.title("Age Feature Before Transformation")
sns.histplot(new_df['age'] , kde=True)
plt.show()
```

```python
new_df = sns.load_dataset('titanic')
trans = PowerTransformer(method="box-cox")

new_df['age'] = trans.fit_transform(new_df[['age']])
plt.title("Age Feature After Box-Cox Transformation")
sns.histplot(new_df['age'] , kde=True)
plt.show()
```

# Quantile Transformation
Quantile transformation is a non-linear transformation technique that maps the original data to a uniform or normal distribution. This method is particularly useful for handling skewed data and outliers, as it can help to stabilize variance and make the data more normally distributed. Quantile transformation works by ranking the data points and then assigning them new values based on their rank, effectively spreading out the values evenly across the desired distribution.

```python
from sklearn.preprocessing import QuantileTransformer

trans = QuantileTransformer(n_quantiles=5)
new_df['fare'] = trans.fit_transform(new_df[['fare']])
plt.title("Fare Feature After Quantile Transformation")
sns.histplot(new_df['fare'] , kde=True)
plt.show()
```

```python
df['age'] = trans.fit_transform(df[['age']])
# df['age'] = trans.inverse_transform(df[['age']])
plt.title("Age After Quantile Transformation")
sns.histplot(df['age'] , kde=True)
plt.show()
```
