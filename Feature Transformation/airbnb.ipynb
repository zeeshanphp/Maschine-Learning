## Feature Transformation in AIRBNB data

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Datasets/airbnb.csv')
df.info()
```

```python
from sklearn.impute import SimpleImputer
impt = SimpleImputer(strategy="most_frequent")
df['name'] = impt.fit_transform(df[['name']]).ravel()
df['name'].isnull().sum()
sns.histplot(df['latitude'] , kde=True)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer

trans = QuantileTransformer(n_quantiles=150)
df['latitude'] = trans.fit_transform(df[['latitude']])
sns.histplot(df['latitude'] , kde=True)
df['latitude'].head()
```
