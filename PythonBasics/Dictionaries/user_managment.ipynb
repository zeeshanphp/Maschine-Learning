```python
users = {}
def add_user(email,name,city):
    users[email] = {"fullname" : name , "adress" : city}


def view_user():
    print("All Users")
    for key, values in users.items():
        print(key, values['fullname'],values['adress'])
    
def search_user(email):
    if email not in users:
        print("User Not Found")
    else:
        print(users[email])
    
def update_user(email,name,city):
    
    users[email] = {"fullname" : name , "adress" : city}
def main():
    print(''' 
        1. Add User \n
        2. View Users \n
        3. Search User \n
        4. Update User \n
        5. Delete User
        ''')
    opt = int(input("Enter Option Number and 0 for Exit"))
    if opt == 0 :
        exit()
    if opt == 1:    
        while True:
            email = input("Enter Email Adress")
            if email.lower() == "no":   
                main()
                break
            name = input("Enter Your Name")
            city = input("Enter Your City")
            add_user(email,name,city)
            main()
    if opt == 2:
        view_user()
        main()
        
    if opt == 3:
        email = input("Enter Email Adress")
        search_user(email)
        main()
        

if __name__ == "__main__":
    main()
```
