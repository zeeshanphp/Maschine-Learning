##### Handling Outlier (Trimming)

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3-Q1
# uppper and lower limits of IQR
upper_limit = Q3 + (IQR*1.5)
lower_limit = Q1 - (IQR*1.5)
upper_limit_outlier = df[df['age'] >= upper_limit]
lower_limit_outlier = df[df['age'] <= lower_limit]
sns.histplot(upper_limit_outlier['age'])
upper_limit_outlier.shape
# Trimming data from upper and lower limit
trimmed_data = df[(df["age"] <= upper_limit) & (df['age'] >= lower_limit)]
# Plotting data on boxplot
# sns.boxplot(trimmed_data['age'])
```

### FARE Column

```python
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3-Q1
uper_limit = Q3 + (IQR*1.5)
lwr_limit = Q1 - (IQR*1.5)
trim_data = df[(df['fare'] <= uper_limit) & (df['fare'] >= lwr_limit)]
# up_outlier = df[(df['fare'] >= uper_limit)]
trim_data.head()
sns.histplot(trim_data['fare'] , kde=True)
```
