```python
df = pd.read_csv('../Datasets/diamonds.csv')

sample = df.sample(700, random_state=42,)
# print(sample.head())
plt.title('Diamonds Dataset Sample')
plt.xlabel('Carat of Diamonds')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
sns.scatterplot(data=sample, x='carat', y='price', hue='cut' , )
```

```python
sns.boxplot(data=sample, x='cut', y='price', hue='cut', palette='Set2' , orient='v')
```

```python
fig = px.bar(data_frame=sample, x='carat', y='price', color='cut', labels={'carat': 'Carat Of Diamonds', 'price': 'Price ($)'}, title='Diamonds Dataset Sample with Plotly' )
fig.show()
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(rows=1, cols=3, subplot_titles=('Scatter Plot', 'Box Plot', 'Bar Plot'))
scatter = px.scatter(data_frame=sample, x='carat', y='price', color='cut' ,).data[0]
fig.add_trace(scatter, row=1, col=1)

box = px.density_heatmap(data_frame=sample, x='carat', y='price').data[0]
fig.add_trace(box, row=1, col=2)

bar = px.bar(data_frame=sample, x='carat', y='price', color='cut').data[0]
fig.add_trace(bar, row=1, col=3)
fig.update_layout(title_text='Diamonds Dataset Sample with Subplots')
fig.show()
```

### ANIMATED PLOTS
For this we have to add **animation_frame ='cut'** in our scatter plot

```python
# Animated Plot
fig = px.scatter(data_frame=sample, x='carat', y='price', color='cut', animation_frame='cut', labels={'carat': 'Carat Of Diamonds', 'price': 'Price ($)'}, title='Diamonds Dataset Sample with Plotly' )
fig.show()
```

### FACET PLOTS
Sada alfaz m ksi b dataset ko categories ki base p divde kr daina r hr category ka different  
plot show ho jae ga jsy hmary dataset m category wise ho ga
You have one dataset (like the diamonds dataset).  
You pick a categorical variable (e.g., category).   
**PLOTLY** will Look at all unique values of that category (e.g., Fair, Good, Very Good, Premium, Ideal).  
Split the dataset into groups, one per unique category.  
Apply the same type of plot (scatter, histogram, box, etc.) to each group.  
Show all those small plots in a grid layout for comparison.  
This whole splitting process is done automatically → you don’t have to filter the dataset   yourself.  



```python
fig = px.scatter(data_frame=sample, x='carat' , y='price' , facet_col='cut', color='cut')
fig.show()
```

### FACET with row and column

```python
fig = px.scatter(data_frame=df, x='carat' , y='price' , facet_col='color' , facet_row='cut', animation_frame='clarity' , color='cut')
fig.update_layout(height=800, width=700)
fig.show()
```
